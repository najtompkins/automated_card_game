# define the functions

```python
def deal():
    # Deal the top cards from the players hands and place them in variables

    # Define the deal result variable
    deal_error = "Deal occurred without error"

    if len(player_1_hand) > 0:
        top1  = player_1_hand.deal(1)
        player_1_war.add(top1)
    else:
        deal_error = "Player 1's hand is empty"

    if len(player_2_hand) > 0:
        top2 = player_2_hand.deal(1)
        player_2_war.add(top2)
    else:
        deal_error = "Player 2's hand is empty"

    return deal_error
    
```

```python
def compare():

    # Define the round winner variable
    round_winner = None

    # assign the top cards in the war stack to varibvles.
    top_card_1 = player_1_war[-1]
    top_card_2 = player_2_war[-1]

    # Compare the cards
    if top_card_1.value > top_card_2.value: # if player 1's card is greater than player 2's
        round_winner = 1
        print(f"Player 1 wins the round with the {top_card_1}!")
    elif top_card_1.value < top_card_2.value: # if player 1's card is less than player 2's
        round_winner = 2
        print(f"Player 2 wins the round with the {top_card_2}!")
    elif top_card_1.value == top_card_2.value: # if player 1's card is equal to player 2's
        round_winner = "war"
        print(f"Neither player wins! It's a war!")
    else: # if something went wrong
        print("Something went wrong!")
        round_winner = 0

    return round_winner
```

```python
def war():

    war_error = "War deals occurred without error"

    # Deal the top 4 cards from the players hands and place them in variables
    for i in range(3):
        deal_result = deal()
        
        if deal_result != "Deal occurred without error":
            # player 1 loses
            if deal_result == 1:
                war_error = "Player 1's hand is empty"
            # player 2 loses
            if deal_result == 2:
                war_error = "Player 2's hand is empty"
```

```python
def reset_war_stacks(result):
    # reset the war stacks

    if result == 1:
        # reset player 1's war stack into player_1_discard
        for i in range(len(player_1_war)):
            temp_card = player_1_war.deal(1)
            player_1_discard.add(temp_card)

        # reset player 2's war stack into player_1_discard
        for i in range(len(player_2_war)):
            temp_card = player_2_war.deal(1)
            player_1_discard.add(temp_card)
    
    elif result == 2:
        # reset player 1's war stack into player_2_discard
        for i in range(len(player_1_war)):
            temp_card = player_1_war.deal(1)
            player_2_discard.add(temp_card)

        # reset player 2's war stack into player_2_discard
        for i in range(len(player_2_war)):
            temp_card = player_2_war.deal(1)
            player_2_discard.add(temp_card)
    
    elif result == 0:
        print("Cannot reset stacks when comparison errors (value 0) are passed into reset function")
    
    elif result == 10:
        print("Hey there's a war going on! Cannot reset the stacks")

    else: print("something went wrong when resetting the stacks.")
```

```python
def check_for_winner(winner):
    if winner == 1 and len(player_2_hand) == 0:
        return 1
    elif winner == 2 and len(player_1_hand) == 0:
        return 2
    else:
        print('There was an error when checking for a winner.')
        return 0
```
