```python
import pydealer as pd
```

```python
# Define a new rank dict, ``new_ranks``, with ranks for card faces only.
war_ranks = {
    "values": {
        "Ace": 13,
        "King": 12,
        "Queen": 11,
        "Jack": 10,
        "10": 9,
        "9": 8,
        "8": 7,
        "7": 6,
        "6": 5,
        "5": 4,
        "4": 3,
        "3": 2,
        "2": 1
    }
}
```

```python
# Set up the deck (automatically built)
starting_deck = pd.Deck()

# Shuffle the deck
starting_deck.shuffle()
```

```python
# Set up the player's hands (automatically empty)
player1_deck = pd.Stack()
player2_deck = pd.Stack()

# test_stack
equal_stack = pd.Stack()
```

```python
# Deal cards from the starting deck (this empties the deck)
dealt_cards1 = starting_deck.deal(26)
dealt_cards2 = starting_deck.deal(26)

# Check that the starting deck has been emptied
print(len(starting_deck))
print(len(dealt_cards1))
print(len(dealt_cards2))
```

```python
# Adds the dealt cards to the palyer's hands
    # ALL CARDS ARE ADDED TO THE END OF THE LIST OF CARDS
    # TOP = END OF LIST = LARGEST INDEX
    # BOTTOM = BEGINNING OF LIST = SMALLEST INDEX
player1_deck.add(dealt_cards1, end="top")
player2_deck.add(dealt_cards2, end="top")
```
